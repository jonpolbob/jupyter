```python
results=[]
resuopcloz=[]

for x in df['animal'].unique().tolist() : #liste des animaux utilises
#for x in ['Animal6'] : #ssai avec un seul animal
        dx = df[df.animal == x]            #select un animal
        for area in range(1,5):           #pour regions 1 a 4
            #ecriture des cumuls de colonne de entct et lardur sur les regions 1 a 4
            
            #CA CA NE MARCHE PAS on vire la lilgne epm avec le Nan qui est present si c'est des reultats epm
            #cnt = dx[dx.an==area and dx.nbtot.isnull()].entct.sum(axis=0)
            
            cnt = dx[dx.an==area].entct.sum(axis=0)
            dur=dx[dx.an==area].lardur.sum(axis=0)
            results.append([x,area,cnt,dur])         
        
        area=0
        opencnt = dx[:].OPtoOP.sum(axis=0)+dx[:].CLtoOP.sum(axis=0)+dx[:].OPtoSLF.sum(axis=0)
        closecnt = dx[:].CLtoCL.sum(axis=0)+dx[:].OPtoCL.sum(axis=0)+dx[:].CLtoSLF.sum(axis=0)
        resuopcloz.append([x,area,opencnt,closecnt])
        
#pour le pivot : on fait un dataframe avec les valeurs et rajoutant des legendes
framepivot = pandas.DataFrame(results, columns=['ani', 'area', 'cnt', 'dur'])
framepivot1 = framepivot.pivot(index='ani', columns='area', values='cnt')
framepivot2 = framepivot.pivot(index='ani', columns='area', values='dur')


print
print 'resultats bras'
print
print resuopcloz

print
print 'resultats individuels'
print
print framepivot

print
print 'count en tableau'
print
print framepivot1

print
print 'duree en tableau'
print
print framepivot2






```
